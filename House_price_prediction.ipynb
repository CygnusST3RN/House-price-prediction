```python
#import all the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the data file

data = pd.read_csv('../Downloads/HousingData.xls')

print(data.info())
```

```python
#first 5 data
data.head()
```

```python
#summary of the data
data.describe()
```

```python
#colummns in the data
data.columns
```

```python
#find the null values
print(data.isnull().sum())
```

```python
#remove the null values
data = data.dropna(axis = 0)
data
```

```python
#divide the data

y = data['MEDV']
x = data.iloc[: , :-1]

print(x.head)
print('-------------------------')
print(y.head)
```

```python
#plot the graphs
plt.plot(y,x)
plt.show()
```

```python
#plot variable by variable 
plt.figure(figsize = (20,15))     #define the dimension of the subplots
plotnumber = 1    #indidcates the plot to plot the graph on
for column in data :
    ax = plt.subplot(3,5,plotnumber)   #number of subplots(3*5 = 15)
    sns.histplot(data[column])
    plt.xlabel(column)
    plotnumber += 1

plt.show()
```

```python
sns.histplot(data)
plt.show()
```

```python
#scatter plot
plt.figure(figsize = (22,15))
plotnumber = 1

for col in data:
    plt.subplot(3,5,plotnumber)
    sns.scatterplot(data[col])
    plt.xlabel(col)
    plotnumber += 1    
    
plt.show()
```

```python
#boxplot
plt.figure(figsize = (20,50))
plotnumber = 1

for col in data:
    plt.subplot(10,2,plotnumber)
    sns.boxplot(data = data, x = col, y = 'MEDV')
    plt.xlabel(col)
    plotnumber += 1

plt.show()
```
